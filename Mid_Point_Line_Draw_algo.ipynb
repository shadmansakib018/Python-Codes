```
def findZone(x0,y0,x1,y1):
  dx=x1-x0
  dy=y1-y0
  if(abs(dx)>abs(dy)):
    if(dx>0 and dy>0):
      zone=0
    elif(dx<0 and dy>0):
      zone=3
    elif(dx<0 and dy<0):
      zone=4
    elif(dx>0 and dy<0):
      zone=7
  else:
    if(dx>0 and dy>0):
      zone=1
    elif(dx<0 and dy>0):
      zone=2
    elif(dx<0 and dy<0):
      zone=5
    elif(dx>0 and dy<0):
      zone=6
  return zone
```

```
def convertToZone0(x,y,zone):
  if(zone==0):
    return x,y
  elif(zone==1):
    x0,y0=y,x
  elif(zone==2):
    x0,y0=y,-x
  elif(zone==3):
    x0,y0=-x,y
  elif(zone==4):
    x0,y0=-x,-y
  elif(zone==5):
    x0,y0=-y,-x
  elif(zone==6):
    x0,y0=-y,x
  elif(zone==7):
    x0,y0=x,-y
  return x0,y0
```

```
def midPointLineDraw(x0,y0,x1,y1):
  lix=pylab.array([x0])
  liy=pylab.array([y0])
  
  dx=x1-x0
  dy=y1-y0
  d=2*dy-dx
  ne=2*(dy-dx)
  e=2*dy
  x=x0
  y=y0
  while(x<x1):
    if(d<=0):#choose E
      d=d+e
      x=x+1
      lix=pylab.append(lix,[x])
      liy=pylab.append(liy,[y])
    else:
      d=d+ne
      x=x+1
      y=y+1
      lix=pylab.append(lix,[x])
      liy=pylab.append(liy,[y])
  return lix,liy

```

```
def convertToOriginalZone(Xlist,Ylist,zone):
  if(zone==0):
    return Xlist,Ylist
  elif(zone==1):
    xlist=Ylist
    ylist=Xlist
  elif(zone==2):
    xlist=-Ylist
    ylist=Xlist
  elif(zone==3):
    xlist=-Xlist
    ylist=Ylist
  elif(zone==4):
    xlist=-Xlist
    ylist=-Ylist
  elif(zone==5):
    xlist=-Ylist
    ylist=-Xlist
  elif(zone==6):
    xlist=Ylist
    ylist=-Xlist
  elif(zone==7):
    xlist=Xlist
    ylist=-Ylist
  return xlist,ylist
```

```
#initialise x0,y0 and x1,y1
x0,y0 = 0,0
x1, y1 = 10,6

zone = findZone(x0,y0,x1,y1)
X0,Y0 = convertToZone0(x0,y0,zone) #returns zone0 points
X1,Y1 = convertToZone0(x1,y1,zone)
Xlist = pylab.array([]) #empty array of x and y cordinates
Ylist = pylab.array([])
Xlist, Ylist = midPointLineDraw(X0,Y0,X1,Y1) #List of Co-ordinates in zone 0
Xlist, Ylist = convertToOriginalZone(Xlist,Ylist,zone) #List of Co-ordinates in original zone
pylab.scatter(Xlist,Ylist,lw=0.5)
pylab.show()
```
