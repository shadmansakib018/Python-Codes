```
def convertToOtherZones(Xlist,Ylist):

  newx=pylab.copy(Xlist)
  newy=pylab.copy(Ylist)

  newx=pylab.append(newx,Ylist)
  newy=pylab.append(newy,Xlist)
  
  newx=pylab.append(newx,Ylist)
  newy=pylab.append(newy,-Xlist)

  newx=pylab.append(newx,Xlist)
  newy=pylab.append(newy,-Ylist)

  newx=pylab.append(newx,-Xlist)
  newy=pylab.append(newy,-Ylist)

  newx=pylab.append(newx,-Ylist)
  newy=pylab.append(newy,-Xlist)

  newx=pylab.append(newx,-Ylist)
  newy=pylab.append(newy,Xlist)

  newx=pylab.append(newx,-Xlist)
  newy=pylab.append(newy,Ylist)
  return newx,newy

```

```
def midPointCircle(radius,centerx,centery):
  d=1-radius
  x=0
  y=radius
  Xlist = pylab.array([x]) #empty array of x and y cordinates
  Ylist = pylab.array([y])

  while(x<y):
    if(d<0):
      d=d+2*x+3
      x=x+1
      Xlist=pylab.append(Xlist,[x])
      Ylist=pylab.append(Ylist,[y])
    else:
      d=d+2*x-2*y+5
      x=x+1
      y=y-1
      Xlist=pylab.append(Xlist,[x])
      Ylist=pylab.append(Ylist,[y])

  xlis,ylis=convertToOtherZones(Xlist,Ylist)

  Xlist=pylab.append(Xlist,xlis)
  Ylist=pylab.append(Ylist,ylis)
  Xlist=Xlist+centerx
  Ylist=Ylist+centery

  return Xlist,Ylist
```

```
radius=30
centerx=10
centery=10
rad=radius/2
diagnolx=m.cos(m.pi/4)*rad
diagnoly=m.sin(m.pi/4)*rad
x1,y1=midPointCircle(radius,centerx,centery)

x2,y2=midPointCircle(rad,centerx+rad,centery)
x3,y3=midPointCircle(rad,centerx-rad,centery)
x4,y4=midPointCircle(rad,centerx,centery+rad)
x5,y5=midPointCircle(rad,centerx,centery-rad)
x6,y6=midPointCircle(rad,centerx+diagnolx,centery+diagnoly)
x7,y7=midPointCircle(rad,centerx-diagnolx,centery+diagnoly)
x8,y8=midPointCircle(rad,centerx-diagnolx,centery-diagnoly)
x9,y9=midPointCircle(rad,centerx+diagnolx,centery-diagnoly)
pylab.rcParams["figure.figsize"] = (10,10)
pylab.scatter(x1,y1)
pylab.scatter(x2,y2)
pylab.scatter(x3,y3)
pylab.scatter(x4,y4)
pylab.scatter(x5,y5)
pylab.scatter(x6,y6)
pylab.scatter(x7,y7)
pylab.scatter(x8,y8)
pylab.scatter(x9,y9)
pylab.show()
```
