```
import math
nf = math.inf
def get():
  return min(fn, key=fn.get)
def goaltest(node):
  if(node==end):
    return True  
def pathfind(node):
  path=""
  while(node!=st):
    path=node+" "+path
    node=parent.get(node)
  return path

f=open('/content/drive/My Drive/Colab Notebooks/A*search/Alpha.txt','r')
g=f.readlines()
line=g[0].split(', ')
st=line[0]
end=line[1][0:-1]
tree={}
h={}
gn={}
for i in range(1,len(g),3):
  li=[]
  line=g[i].split(', ')
  line[-1]=line[-1][0:-1]
  line2=g[i+1].split(',')
  line2[-1]=line2[-1][0:-1]
  line3=g[i+2]
  line3=line3[0:-1]
  node_current=line[0]
  for i in range(len(line2)):
    li.append([line[i+1] , int(line2[i])])
  tree.update({node_current : li})
  h.update({node_current: int(line3)})
  gn.update({node_current:nf})
#print(tree)
#print(h)
#print(gn)
parent={}
fn={}

parent.update({st:st})
gn.update({st:0})
fn.update({st:h.get(st)})
frontier=[]
explored=[]
frontier.append(st)

while(len(frontier)):
  current=get()
  frontier.remove(current)
  fn.pop(current)
  if(goaltest(current)):
    str=st+ " " + pathfind(current)
    print("The shortest path to destination: "+str+". Total cost: ",gn.get(current))
    break
    
  else:
    for i in tree[current]:
      #print(current)
      gc=gn.get(current)+i[1]
      if(gc<(gn.get(i[0]))):
        parent.update({i[0]:current})
        gn.update({i[0]:gc})
        fn.update({i[0]:(gn.get(i[0])+h.get(i[0]))})
        if(frontier.count(i[0])==0):
          frontier.append(i[0])
```
